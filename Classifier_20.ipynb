```python

def fit(files,vocab):
    dictionary = {}
    for ok in files:
        #dictionary[i[i.index('\\')+1:]]={} make it at last  
        
        str1=ok+'/*'
        
        words=[]
        
        f1=glob.glob(str1)
        for j in f1:
            with open(j,'r') as f:
                temp=f.read()
                temp=temp[temp.index('Date:'):]
            
                for word in temp.split():
                    a=word.lower()
                    a=a.translate(table)
                    if(a[len(a)-1:]=='s'):
                        a=a[:len(a)-1]
                    words.append(a)
                    
        di={}
        for i in words:
            di[i]=1
        for i in words:
            if(di[i]>=1):
                di[i]+=1 
                
#         for i in di:
#             if(di[i]<2500 and di[i]>40 and len(i)>3): necessary ??? i don't think so 
#                 temp[i]=di[i]
        class_final={}
        for i in vocab.keys():
            if(i not in di.keys() or di[i]==0):
                class_final[i]=1
            else:
                class_final[i]=di[i]
            
        dictionary[ok[ok.index('\\')+1:]]=class_final.copy()#made it at last
        
        s=0
        for i in class_final:
            s+=vocab[i]
            
        dictionary[ok[ok.index('\\')+1:]]['total_count']=s
        
    return dictionary         
                
    
```

```python
def probability(dictionary,words,current_class):
    s=0
    for i in words:
        if(i in dictionary[current_class].keys()):
            s+= np.log(dictionary[current_class][i]/dictionary[current_class]['total_count'])
    return s    
```

```python

def predict(files,dictionary) :
    ans=[]
    for ok in files:
        words=[]
        with open(ok,'r') as f:
            temp=f.read()
            temp=temp[temp.index('Date:'):]
            for word in temp.split():
                a=word.lower()
                a=a.translate(table)
                if(a[len(a)-1:]=='s'):
                    a=a[:len(a)-1]
                words.append(a)
        classes=dictionary.keys()   
        
        best_p=-1
        best_class='none'
        first_run=True
        
        for current_class in classes:
            p_current_class = probability(dictionary, words, current_class)
            
            if (first_run or p_current_class > best_p):
                best_p = p_current_class
                best_class = current_class
            first_run = False
        ans.append(best_class)
    
    return ans
```

```python
files2=glob.glob("news_train/*")
total=0

ytest=[]
ypred=[]
for ok2 in files2:
    idk=glob.glob(ok2+'/*')
    comp=ok2[11:]
    ans=predict(idk,checkit)
    
    count=0
    for a in ans:
        ytest.append(comp)
        ypred.append(a)
        if(a==comp):
            count+=1
    total+=count      
    print(count,'/',200,':', comp)
print (total)
```
